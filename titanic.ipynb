```python
one_hoted_dataset = pd.get_dummies(dataset["Embarked"],dtype=int)
```

```python
new_dataset = pd.concat([one_hoted_dataset,new_dataset], axis=1)
```

```python
new_dataset
```

```python
del  new_dataset["Sex"]
del  new_dataset["Embarked"]
del  new_dataset["Cabin"]
```

```python
new_dataset.isnull().sum()
```

```python
age_mean = new_dataset["Age"].mean()
new_dataset["Age"].fillna(age_mean,inplace=True)
```

```python
new_dataset["Nsurvived"]=new_dataset["Survived"]
```

```python
del new_dataset["Survived"]
```

```python
features = new_dataset.iloc[:, :-1]
labels = new_dataset.iloc[:, -1]
from sklearn.model_selection import train_test_split
x_tr, x_ts, y_tr, y_ts = train_test_split(features,labels, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler=StandardScaler()
x_tr=scaler.fit_transform(x_tr)
x_ts=scaler.transform(x_ts)
```

************ Modelling ****************

Disition Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
tree = DecisionTreeClassifier()
tree.fit(x_tr,y_tr)
pred = tree.predict(x_ts)
print(accuracy_score(pred, y_ts))
print(confusion_matrix(pred, y_ts))
```

SVM

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(x_tr, y_tr)
y_pred = svm.predict(x_ts)
print(accuracy_score(y_pred, y_ts))
print(confusion_matrix(y_pred, y_ts))
```

Neural Network

```python
from sklearn.neural_network import MLPClassifier
model = MLPClassifier(hidden_layer_sizes=(9,8,5))
model.fit(x_tr, y_tr)
predict = model.predict(x_ts)
print(accuracy_score(y_ts,predict))
print(confusion_matrix(y_ts,predict))
```
